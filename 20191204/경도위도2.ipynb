```python
import pandas as pd
```

```python
df = pd.read_csv('bus.csv')
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['총통행거리']=df['총통행거리']*0.001
```

```python
df['총통행거리-직선거리']=df['총통행거리']-df['직선거리']
```

```python
df.sort_values(by=['총통행거리-직선거리'],inplace=True,ascending=False)
```

```python
color = ['blue', 'orange', 'red', 'darkblue', 'cadetblue', 'gray', 'lightgreen', 'lightblue', 'purple', 'green']
df['색']=color
```

```python
import folium
import folium.plugins
```

```python
m = folium.Map(location=[37.2080477, 127.0204461],zoom_start=13)
fg = folium.FeatureGroup(name = '승차정류장')
for lat, lng, name,col in zip(df['WGS84위도_승차'], df['WGS84경도_승차'], df['정류소명_승차'],df['색']):
    fg.add_child(folium.Marker(
        location = [lat, lng],
        popup= (folium.Popup(name)),
        icon = folium.Icon(color=col)
    ))

fg2 = folium.FeatureGroup(name = '하차정류장')
for lat, lng, name,col in zip(df['WGS84위도_하차'], df['WGS84경도_하차'], df['정류소명_하차'],df['색']):
    fg2.add_child(folium.Marker(
        location = [lat, lng],
        popup= (folium.Popup(name)),
        icon = folium.Icon(color=col)
    ))

busline=[]
for lat1,lng1,lat2,lng2,col in zip(df['WGS84위도_승차'], df['WGS84경도_승차'],df['WGS84위도_하차'], df['WGS84경도_하차'], df['색']):
    busline.append([[lat1,lng1],[lat2,lng2]])

i=0
for col in df['색']:  
    folium.PolyLine(busline[i], color=col, weight=5, opacity=1).add_to(m)
    i+=1
    
m.add_child(fg)        
m.add_child(fg2)
```
