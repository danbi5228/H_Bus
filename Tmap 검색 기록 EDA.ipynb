```python
Tmap_map = pd.read_csv('Tmap_coordinates.csv')
```

```python
Tmap_map.sort_values(by= 'score')
```

```python
for i in Tmap_map['score']:
    if i <= 100 :
        Tmap_map['score'].replace(i, 100, inplace = True)
```

```python
map = folium.Map(location=[37.1995, 126.8312], zoom_start=11)

fg1 = folium.FeatureGroup(name = 'Tmap')
for lat, lng, name, score in zip(Tmap_map['위도'], Tmap_map['경도'], Tmap_map['검색지명'], Tmap_map['score']):
    fg1.add_child(folium.Circle(
        location=[lat, lng],
        popup = (folium.Popup(name)),
        radius= score,
        weight=1,
        opacity=1,
        color = '#8000ff',
        fill_color='#8000ff',
        fill_opacity=0.5
    ))
map.add_child(fg1)


map
```
